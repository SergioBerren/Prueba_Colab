**INTRODUCCIÓN AL MACHINE LEARNING CON UNA PREDICCIÓN SIMPLE**

CASO PRÁCTICO: VENTAS DE HELADOS SEGÚN LA TEMPERATURA

*OBJETIVO: ENTENDER CÓMO LOS DATOS PUEDEN USARSE PARA REALIZAR PREDICCIONES SIMPLES*

IMPORTAR LIBRERÍAS

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

CARGAR EL DATASET

Columnas:
- day: fecha
- temperature_c: termperatura en celsius
- icecream_sales: helados vendidos

```python
df = pd.read_csv('icecream_sales_temperature.csv')
df.head()
```

EXPLORACIÓN DEL DATASET

```python
df.info()
```

```python
df.describe()
```

LIMPIEZA BÁSICA

```python
df = df.drop_duplicates()
df = df.dropna()
```

VISUALIZACIÓN 1: DISTRIBUCIÓN DE TEMPERATURAS

```python
plt.hist(df['temperature_c'])
plt.title('Distribución de temperaturas')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia')
plt.show()

# Interpretación: muestra qué temperaturas aparecen más.
```

VISUALIZACIÓN 2: RELACIÓN TEMPERATURA VS VENTAS

```python
plt.scatter(df['temperature_c'], df['icecream_sales'])
plt.title('Temperatura vs Ventas de Helados')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Helados vendidos')
plt.show()

# Interpretación: cuando la temperatura aumenta, las ventas también.
```

PREPARAR LOS DATOS

```python
X = df[['temperature_c']]
y = df['icecream_sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

ENTRENAR EL MODELO DE REGRESIÓN LINEAL

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

VISUALIZACIÓN 3: LÍNEA DE REGRESIÓN

```python
pred = model.predict(X)
plt.scatter(df['temperature_c'], df['icecream_sales'])
plt.plot(df['temperature_c'], pred)
plt.title('Regresión lineal: temperatura vs ventas')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Helados vendidos')
plt.show()

# Interpretación: la línea muestra la relación que el modelo ha aprendido.
```

PREDICCIÓN

```python
prediction = model.predict([[30]])
print('Helados estimados vendidos a 30°C:', prediction)
```
